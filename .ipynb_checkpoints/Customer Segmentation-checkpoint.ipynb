```python
df = df.na.drop()
```

```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
```

```python
# 1. Segmentation using Age and Spending score
```

```python
assemble = VectorAssembler(inputCols=[ 'Age' ,'Spending Score (1-100)'], outputCol='features')

'''
    In PySpark we have to create a separate column known as
    'features' which is a vector form of all features.
''';

assembled_data = assemble.transform(df)
assembled_data.show(2)
```

```python
scale = StandardScaler(inputCol='features', outputCol='standardized')

'''
    The models tend to perform better when we have the features
    standardized to a specific scale.
''';

'''
    To standardize the data to bring them to a comparable scale
    Euclidean distance is always impacted more by variables on a higher scale,
    hence itâ€™s important to scale the variables out.
''';

data_scale = scale.fit(assembled_data)

data_scale_output = data_scale.transform(assembled_data)
data_scale_output.show(2)
```

```python
# Create a evaluator to evaluate the clustering performance

evaluator = ClusteringEvaluator(
    predictionCol='prediction',
    featuresCol='standardized',
    metricName='silhouette',
    distanceMeasure='squaredEuclidean'
)

'''
    Silhouette Score using ClusteringEvaluator() measures how close
    each point in one cluster is to points in the neighboring clusters
    thus helping in figuring out clusters that are compact and well-spaced out.
''';

silhouette_score = []

print("""
Silhoutte Scores for K Mean Clustering
======================================
Model\tScore\t
=====\t=====\t
""")

for k in range(2,11):
    kmeans_algo = KMeans(featuresCol='standardized', k=k)
    kmeans_fit = kmeans_algo.fit(data_scale_output)

    output = kmeans_fit.transform(data_scale_output)
    score=evaluator.evaluate(output)

    silhouette_score.append(score)
    print(f"K{k}\t{round(score,2)}\t")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.plot(range(2,11), silhouette_score)
ax.set_xlabel("K")
ax.set_ylabel("Score");
```

```python
'''
    The optimal number of clusters was found to be '6',
    when the segmentation factors were Age and Spending score
''';
```

```python
# 2. Segmentation using Annual income and Spending score
```

```python
assemble = VectorAssembler(inputCols=['Annual Income (k$)', 'Spending Score (1-100)'], outputCol='features')

assembled_data = assemble.transform(df)
assembled_data.show(2)
```

```python
data_scale = scale.fit(assembled_data)

data_scale_output = data_scale.transform(assembled_data)
data_scale_output.show(2)
```

```python
silhouette_score = []

print("""
Silhoutte Scores for K Mean Clustering
======================================
Model\tScore\t
=====\t=====\t
""")

for k in range(2,11):
    kmeans_algo = KMeans(featuresCol='standardized', k=k)
    kmeans_fit = kmeans_algo.fit(data_scale_output)

    output = kmeans_fit.transform(data_scale_output)
    score=evaluator.evaluate(output)

    silhouette_score.append(score)
    print(f"K{k}\t{round(score,2)}\t")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.plot(range(2,11), silhouette_score)
ax.set_xlabel("K")
ax.set_ylabel("Score");
```

```python
'''
    The optimal number of clusters was found to be '5',
    when the segmentation factors were Annual Income and Spending Score.
''';
```

```python
# 3. Segmentation with all the 3 features (Age, Annual Income and Spendin score)
```

```python
assemble = VectorAssembler(inputCols=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'], outputCol='features')

assembled_data = assemble.transform(df)
assembled_data.show(2)
```

```python
data_scale = scale.fit(assembled_data)

data_scale_output = data_scale.transform(assembled_data)
data_scale_output.show(2)
```

```python
silhouette_score = []

print("""
Silhoutte Scores for K Mean Clustering
======================================
Model\tScore\t
=====\t=====\t
""")

for k in range(2,11):
  kmeans_algo = KMeans(featuresCol='standardized', k=k)
  kmeans_fit = kmeans_algo.fit(data_scale_output)
  output = kmeans_fit.transform(data_scale_output)
  score=evaluator.evaluate(output)
    
  silhouette_score.append(score)

  print(f"K{k}\t{round(score,2)}\t")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.plot(range(2,11), silhouette_score)
ax.set_xlabel("K")
ax.set_ylabel("Score");
```

```python
'''
    The optimal number of clusters was found to be '6',
    when the segmentation factors were Age, Annual Income and Spending Score.
''';
```
